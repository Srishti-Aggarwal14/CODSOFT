```python
# 📱 Welcome to Srishti's Contact Keeper 📱

my_contacts = {}

def display_menu():
    print("\n🔖 My Contact Book")
    print("1️⃣ Add New Contact")
    print("2️⃣ View All Contacts")
    print("3️⃣ Search by Name")
    print("4️⃣ Edit Existing Contact")
    print("5️⃣ Remove Contact")
    print("6️⃣ Exit")

def add_new_contact():
    name = input("👤 Name: ").strip()
    phone = input("📞 Phone Number: ").strip()
    email = input("📧 Email: ").strip()
    address = input("🏠 Address: ").strip()

    my_contacts[name] = {
        'Phone': phone,
        'Email': email,
        'Address': address
    }

    print(f"✅ '{name}' has been added successfully!")

def show_contacts():
    if not my_contacts:
        print("📭 No contacts saved yet.")
    else:
        print("📋 Your Saved Contacts:")
        for name, info in my_contacts.items():
            print(f"\n🔸 Name: {name}")
            print(f"   📞 Phone: {info['Phone']}")
            print(f"   📧 Email: {info['Email']}")
            print(f"   🏠 Address: {info['Address']}")

def find_contact():
    search_name = input("🔍 Enter name to search: ").strip()
    contact = my_contacts.get(search_name)

    if contact:
        print(f"\n👤 Contact: {search_name}")
        print(f"📞 {contact['Phone']}")
        print(f"📧 {contact['Email']}")
        print(f"🏠 {contact['Address']}")
    else:
        print(f"❌ '{search_name}' not found in contacts.")

def update_contact():
    name = input("✏️ Enter name to edit: ").strip()
    if name in my_contacts:
        print("Leave field blank to keep current value.")
        new_phone = input(f"📞 New phone (current: {my_contacts[name]['Phone']}): ").strip()
        new_email = input(f"📧 New email (current: {my_contacts[name]['Email']}): ").strip()
        new_address = input(f"🏠 New address (current: {my_contacts[name]['Address']}): ").strip()

        if new_phone: my_contacts[name]['Phone'] = new_phone
        if new_email: my_contacts[name]['Email'] = new_email
        if new_address: my_contacts[name]['Address'] = new_address

        print(f"✅ '{name}' updated successfully!")
    else:
        print("❌ No such contact found.")

def delete_contact():
    name = input("🗑️ Enter name to delete: ").strip()
    if name in my_contacts:
        del my_contacts[name]
        print(f"🗑️ '{name}' deleted from contacts.")
    else:
        print("❌ Contact not found.")

# 🎉 Start the Application
print("📱 Welcome to Srishti's Smart Contact Book!")

while True:
    display_menu()
    choice = input("🔸 Choose an option (1–6): ")

    if choice == '1':
        add_new_contact()
    elif choice == '2':
        show_contacts()
    elif choice == '3':
        find_contact()
    elif choice == '4':
        update_contact()
    elif choice == '5':
        delete_contact()
    elif choice == '6':
        print("👋 Exiting... Stay connected! 💫")
        break
    else:
        print("⚠️ Invalid option. Please choose between 1 and 6.")
```
