```python
# ğŸ“± Welcome to Srishti's Contact Keeper ğŸ“±

my_contacts = {}

def display_menu():
    print("\nğŸ”– My Contact Book")
    print("1ï¸âƒ£ Add New Contact")
    print("2ï¸âƒ£ View All Contacts")
    print("3ï¸âƒ£ Search by Name")
    print("4ï¸âƒ£ Edit Existing Contact")
    print("5ï¸âƒ£ Remove Contact")
    print("6ï¸âƒ£ Exit")

def add_new_contact():
    name = input("ğŸ‘¤ Name: ").strip()
    phone = input("ğŸ“ Phone Number: ").strip()
    email = input("ğŸ“§ Email: ").strip()
    address = input("ğŸ  Address: ").strip()

    my_contacts[name] = {
        'Phone': phone,
        'Email': email,
        'Address': address
    }

    print(f"âœ… '{name}' has been added successfully!")

def show_contacts():
    if not my_contacts:
        print("ğŸ“­ No contacts saved yet.")
    else:
        print("ğŸ“‹ Your Saved Contacts:")
        for name, info in my_contacts.items():
            print(f"\nğŸ”¸ Name: {name}")
            print(f"   ğŸ“ Phone: {info['Phone']}")
            print(f"   ğŸ“§ Email: {info['Email']}")
            print(f"   ğŸ  Address: {info['Address']}")

def find_contact():
    search_name = input("ğŸ” Enter name to search: ").strip()
    contact = my_contacts.get(search_name)

    if contact:
        print(f"\nğŸ‘¤ Contact: {search_name}")
        print(f"ğŸ“ {contact['Phone']}")
        print(f"ğŸ“§ {contact['Email']}")
        print(f"ğŸ  {contact['Address']}")
    else:
        print(f"âŒ '{search_name}' not found in contacts.")

def update_contact():
    name = input("âœï¸ Enter name to edit: ").strip()
    if name in my_contacts:
        print("Leave field blank to keep current value.")
        new_phone = input(f"ğŸ“ New phone (current: {my_contacts[name]['Phone']}): ").strip()
        new_email = input(f"ğŸ“§ New email (current: {my_contacts[name]['Email']}): ").strip()
        new_address = input(f"ğŸ  New address (current: {my_contacts[name]['Address']}): ").strip()

        if new_phone: my_contacts[name]['Phone'] = new_phone
        if new_email: my_contacts[name]['Email'] = new_email
        if new_address: my_contacts[name]['Address'] = new_address

        print(f"âœ… '{name}' updated successfully!")
    else:
        print("âŒ No such contact found.")

def delete_contact():
    name = input("ğŸ—‘ï¸ Enter name to delete: ").strip()
    if name in my_contacts:
        del my_contacts[name]
        print(f"ğŸ—‘ï¸ '{name}' deleted from contacts.")
    else:
        print("âŒ Contact not found.")

# ğŸ‰ Start the Application
print("ğŸ“± Welcome to Srishti's Smart Contact Book!")

while True:
    display_menu()
    choice = input("ğŸ”¸ Choose an option (1â€“6): ")

    if choice == '1':
        add_new_contact()
    elif choice == '2':
        show_contacts()
    elif choice == '3':
        find_contact()
    elif choice == '4':
        update_contact()
    elif choice == '5':
        delete_contact()
    elif choice == '6':
        print("ğŸ‘‹ Exiting... Stay connected! ğŸ’«")
        break
    else:
        print("âš ï¸ Invalid option. Please choose between 1 and 6.")
```
